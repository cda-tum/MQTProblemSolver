# Import and Problem Visualization

```python
from tsp import TSP
tsp = TSP()
tsp.print()
```

# Definition of Problem Instance

```python
dist_1_2=6
dist_1_3=9
dist_1_4=2
dist_2_3=1
dist_2_4=8
dist_3_4=4
```

# Get available quantum algorithms

```python
available_quantum_algorithms = tsp.get_available_quantum_algorithms()
available_quantum_algorithms
```

# Solve the problem accordingly

```python
for algo in available_quantum_algorithms:
    print("Algorithm: ", algo)
    res = tsp.solve(dist_1_2=dist_1_2,dist_1_3=dist_1_3,dist_1_4=dist_1_4,dist_2_3=dist_2_3,dist_2_4=dist_2_4,dist_3_4=dist_3_4, quantum_algorithm=algo, objective_function="shortest_path", num_qubits_qft=8)
    print(res)
```

# Visualize solution

```python
tsp.print(solution=res)
```

# Show classical solution

```python
tsp.show_classical_solution()
```
